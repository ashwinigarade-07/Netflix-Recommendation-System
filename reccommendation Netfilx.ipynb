```python
df_select = df[['Title','Description','Genres','Content Type']]
df_select
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
df_select["Title"] = df_select["Title"].apply(clean)
```

```python
df_select
```

```python
#Now I will use the Genres column as the feature to recommend similar content to the user. 
#I will use the concept of cosine similarity here (used to find similarities in two documents)
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df['combine_feature'] = df['Title']+" " +df['Description']+ " "+df['Genres']+ " "+df['Content Type']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combine_feature'])
```

```python
#cosine similairty
cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
cosine_sim
```

```python
def recommend(movie_name, cosine_sim=cosine_sim):
    movie_name = movie_name.lower()
    
    df['Title_lower'] = df['Title'].str.lower()
    
    # Find movies that contain the word
    matches = df[df['Title_lower'].str.contains(movie_name)]

    if matches.empty:
        return "Movie not found"
    
    idx = matches.index[0]   # take first match
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    
    return df['Title'].iloc[movie_indices]
```

```python
print(recommend('alive'))
```

```python
pickle.dump(df, open("netflix_df.pkl", "wb"))
pickle.dump(cosine_sim, open("cosine_sim.pkl", "wb"))
pickle.dump(tfidf, open("tfidf.pkl", "wb"))

print("Pickle files saved successfully!")
```
